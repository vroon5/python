## ER miner Algorithm

```python
# reference for all SMPF algorithms: https://www.philippe-fournier-viger.com/spmf/documentation.php
# Github repository for original wrapper code: https://github.com/fandu/maximal-sequential-patterns-mining
# Reference document for the algorithm below: http://www.philippe-fournier-viger.com/spmf/ERMiner.php
# Other reference: http://data-mining.philippe-fournier-viger.com/introduction-to-sequential-rule-mining/


# $$$$$$$$$$$$$ ER miner Algorithm $$$$$$$$$$$$$$$$$$$$$$$$$$$$
 
max_pattern_length=2
max_leng_bet_seq=2
show_seq_id="true" 
algorithm="ERMiner" #"VMSP"  
min_confidence=0.45
import subprocess
import re

class erminer:
    def __init__(self):
        self._executable = "spmf.jar"
        self._input = "input_java.txt"
        self._output = "output.txt"

    def run(self, min_supp=0.10):
        # algorithm call
        #java -jar spmf.jar run ERMiner contextPrefixSpan.txt output.txt 75% 50%       
        subprocess.call(["java", "-jar", self._executable, "run", algorithm, self._input, self._output, str(min_supp),str(min_confidence)])
        #       str(max_pattern_length),str(max_leng_bet_seq),show_seq_id

    def encode_input(self, data):
        pass

    def decode_output(self):
        # read
        lines = []
        try:
            with open(self._output, "rU") as f:
                lines = f.readlines()
        except:
            print ("read_output error")
        
        # decode
        patterns = []
        for line in lines:
            line = line.strip()
            patterns.append(list(filter(None,re.split(' #| ==> |SUP: |CONF: ',line))))
# ER miner split ' #| ==> |SUP: |CONF: '
        return patterns


if __name__ == "__main__":
    erminer = erminer()
    erminer.encode_input([])
    erminer.run()

print(erminer.decode_output())
```

```python
# ER miner output
# Pre: precedent Anti: anticedent support: p(A->B) confidence: p(A->B)/A (similar to P(B|A))

erminer_df=pd.DataFrame(erminer.decode_output(),columns=['Pre','Anti','Support','Confidence']).sort_values(['Support','Confidence'],ascending=False)
erminer_df["%support"]=erminer_df.Support.astype(int)/len(data.index)
erminer_df
```

## Trule Algorithm

```python
# reference for all SMPF algorithms: https://www.philippe-fournier-viger.com/spmf/documentation.php
# Reference document for the algorithm below: http://www.philippe-fournier-viger.com/spmf/TRuleGrowth.php
# Other reference: http://data-mining.philippe-fournier-viger.com/introduction-to-sequential-rule-mining/

# $$$$$$$$$$$$$ Trule growth for limiting the window of occurence $$$$$$$$$$$$$$$$$$$$$$$$$

window=3
show_seq_id="true" 
algorithm="TRuleGrowth" 
min_confidence=0.2
import subprocess
import re

class trule:
    def __init__(self):
        self._executable = "spmf.jar"
        self._input = "input_java.txt"
        self._output = "output.txt"

    def run(self, min_supp=0.1):
        # Algorithm call
        # java -jar spmf.jar run TRuleGrowth contextPrefixSpan.txt output.txt 70% 80% 3
        subprocess.call(["java", "-jar", self._executable, "run", algorithm, self._input, self._output, str(min_supp),str(min_confidence),str(window)])
        #       str(max_pattern_length),str(max_leng_bet_seq),show_seq_id

    def encode_input(self, data):
        pass

    def decode_output(self):
        # read
        lines = []
        try:
            with open(self._output, "rU") as f:
                lines = f.readlines()
        except:
            print ("read_output error")

        # decode
        patterns = []
        for line in lines:
            line = line.strip()
            patterns.append(list(filter(None,re.split(' #| ==> |SUP: |CONF: ',line))))

        return patterns


if __name__ == "__main__":
    trule = trule()
    trule.encode_input([])
    trule.run()

print(trule.decode_output())
```

```python
# Trule miner output
# Pre: precedent Anti: anticedent support: p(A->B) confidence: p(A->B)/A (similar to P(B|A))

trule_df=pd.DataFrame(trule.decode_output(),columns=['Pre','Anti','Support','Confidence']).sort_values(['Pre','Confidence','Support'],ascending=False)
trule_df["%support"]=trule_df.Support.astype(int)/len(data.index)
pd.set_option('display.max_rows', None)
print(trule_df[trule_df['Pre']=='12'])
# trule_df[trule_df['Pre']=='3']
trule_df
```

```python
# VMSP from original file
# Github repository for original wrapper code: https://github.com/fandu/maximal-sequential-patterns-mining

import subprocess


class Vmsp:

    def __init__(self):

        self._executable = "spmf.jar"

        self._input = "input_java.txt"

        self._output = "output.txt"



    def run(self, min_supp=0.15):

        # java -jar spmf.jar run VMSP contextPrefixSpan.txt output.txt 50%

        subprocess.call(["java", "-jar", self._executable, "run", "VMSP", self._input, self._output, str(min_supp)])



    def encode_input(self, data):

        pass



    def decode_output(self):

        # read

        lines = []

        try:

            with open(self._output, "rU") as f:

                lines = f.readlines()

        except:

            print ("read_output error")



        # decode

        patterns = []

        for line in lines:

            line = line.strip()

            patterns.append(line.split(" -1 "))



        return patterns





if __name__ == "__main__":

    vmsp = Vmsp()

    vmsp.encode_input([])

    vmsp.run()

    print (vmsp.decode_output())
```
