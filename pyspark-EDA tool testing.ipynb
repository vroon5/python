```python
vhsp_nt_nul_rwnm=sqlContext.sql("SELECT asset,vehicle_Speed,time_ms,row_number() OVER(PARTITION BY asset,date(time_ms) ORDER BY\
asset,date(time_ms)) as rwnm FROM vhsp_nt_nult")
vhsp_nt_nul_rwnmt=vhsp_nt_nul_rwnm.registerTempTable("vhsp_nt_nul_rwnmt")
```

```python
vhsp_nt_nul_rwnm_bet=sqlContext.sql("SELECT a.asset,a.vehicle_speed,a.time_ms AS time_low,b.time_ms AS time_high,a.rwnm,b.rwnm FROM \
vhsp_nt_nul_rwnmt AS a INNER JOIN vhsp_nt_nul_rwnmt AS b ON a.asset=b.asset AND date(a.time_ms)=date(b.time_ms) AND \
b.rwnm=a.rwnm+1 ORDER BY a.rwnm LIMIT 110")
vhsp_nt_nul_rwnm_bett=vhsp_nt_nul_rwnm_bet.registerTempTable("vhsp_nt_nul_rwnm_bett")
```
